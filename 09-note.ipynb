```python
import random

while True:
    rand_num = random.randint(1,6)
    print(rand_num)
    if rand_num == 4:
        break
    print('.')
```

breakは止まる。continueはループする

```python
import random

while True:
    rand_num = random.randint(1,6)
    print(rand_num)
    if rand_num == 4:
        continue
print('.')
```
